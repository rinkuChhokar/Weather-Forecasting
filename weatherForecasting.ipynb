```python
# First lets fetch our dataset from github

import urllib.request
from google.colab import files

# download the file from GitHub
url = "https://raw.githubusercontent.com/rinkuChhokar/MachineLearningModels/main/DailyDelhiClimateTrain.csv"
filename = "DailyDelhiClimateTrain.csv"
urllib.request.urlretrieve(url, filename)

# upload the file to Google Colab, [Note:- click on "Cancel Upload" button when file show on sidebar.]
files.upload()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

data = pd.read_csv("DailyDelhiClimateTrain.csv")
print(data.head())
```

```python
print(data.describe())
```

```python
print(data.info())
```

```python
fig = px.line(data, x="date",
        y="meantemp",
        title="Mean Temperature in Delhi Over The Years")

fig.show()
```

```python
fig = px.line(data, x="date",
        y="humidity",
        title="Humidity in Delhi Over The Years")

fig.show()
```

```python
fig = px.line(data, x="date",
        y="wind_speed",
        title="Wind speed in Delhi Over The Years")

fig.show()
```

```python
fig = px.scatter(data_frame=data, x="humidity",
        y="meantemp", size="meantemp",
        title="Relationship Between Temperature and Humidity in Delhi Over The Years")

fig.show()
```

```python
data["date"] = pd.to_datetime(data["date"], format = '%Y-%m-%d')
data['year'] = data['date'].dt.year
data["month"] = data["date"].dt.month
print(data.head())
```

```python
plt.style.use("fivethirtyeight")
plt.figure(figsize=(15, 10))
plt.title("Temperature Change in Delhi Over the Years")
sns.lineplot(data = data, x="month", y="meantemp", hue="year")
plt.show()
```

```python
forecast_data = data.rename(columns = {"date":"ds",
                       "meantemp":"y"})
print(forecast_data)
```

```python
# facebook prophet model-

from prophet import Prophet
from prophet.plot import plot_plotly, plot_components_plotly
model = Prophet()
model.fit(forecast_data)
forecasts = model.make_future_dataframe(periods=365)
predictions = model.predict(forecasts)
plot_plotly(model, predictions)
```
